## 3. Generating Pointmap from Depth Image 

```python
def extract_intrinsics_safe(yaml_path):
    # 1. Read the file as raw text to avoid ScannerError
    with open(yaml_path, 'r') as f:
        lines = f.readlines()

    # 2. Clean and parse manually into a dictionary
    config = {}
    for line in lines:
        # Strip comments and whitespace
        line = line.split('#')[0].strip()
        if not line or ':' not in line:
            continue
            
        # Split by the FIRST colon only
        key, value = line.split(':', 1)
        config[key.strip()] = float(value.strip())

    # 3. Build the 3x3 Intrinsic Matrix K
    # Using the keys exactly as they appear in your oxford.yaml
    K = np.array([
        [config['Camera.fx'], 0.0,               config['Camera.cx']],
        [0.0,               config['Camera.fy'], config['Camera.cy']],
        [0.0,               0.0,               1.0]
    ], dtype=np.float64)

    # 4. Build Distortion Coefficients
    dist_coeffs = np.array([
        config['Camera.k1'],
        config['Camera.k2'],
        config['Camera.p1'],
        config['Camera.p2'],
        config.get('Camera.k3', 0.0)
    ], dtype=np.float64)

    width = int(config['Camera.width'])
    height = int(config['Camera.height'])

    return K, dist_coeffs, width, height

# Usage
K, distortion, W, H = extract_intrinsics_safe(f'{PATH}/../tracking_dataset/omd/oxford.yaml')
```

```python
import cv2
depth_path = f"{PATH}/../tracking_dataset/omd/depth/000001.png"
# depth = iio.imread(depth_path).astype(np.uint16)
# depth = depth / 1000.0 #convert to mm -> m
# depth[depth <= 0] = np.nan  
# Z = depth

# Constants from oxford.yaml file
baseline = 0.20802998918481966
DepthMapFactor = 256.0
disp_raw = cv2.imread(depth_path, cv2.IMREAD_UNCHANGED).astype(np.float32)
disparity = disp_raw / DepthMapFactor

fx = K[0, 0]
fy = K[1, 1]
cx = K[0, 2]
cy = K[1, 2]

u = np.arange(W)
v = np.arange(H)
uu, vv = np.meshgrid(u, v)

Z = (baseline * fx) / np.where(disparity > 0, disparity, np.nan)
X = (uu - cx) * Z / fx
Y = (vv - cy) * Z / fy

# Convert to right-handed PyTorch3D coordinates
pointmap = np.stack([-X, -Y, Z], axis=-1)
pointmaP = torch.tensor(pointmap, dtype=torch.float32)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Set labels
ax.set_xlabel('X (Horizontal)')
ax.set_ylabel('Y (Vertical)')
ax.set_zlabel('Z (Depth)')
ax.set_title('SAM3D Masked Object Reconstruction')
colors = plt.cm.get_cmap('tab10', 4)

for i, binary_mask in enumerate(masks):
    binary_mask = torch.from_numpy(binary_mask)
    print(binary_mask.shape)
    y_indices, x_indices = torch.where(binary_mask > 0)
    # y_indices = y_indices.to(points.device)
    # x_indices = x_indices.to(points.device)
    filtered_points = pointmap[y_indices, x_indices]
    pts = filtered_points[::50]

    xs, ys, zs = pts[:, 0], pts[:, 1], pts[:, 2]
    img = ax.scatter(xs, ys, zs, label=f'Object ID: {i+1}', c=colors(i), s=5, alpha=0.8)
    
ax.legend()
plt.show()
```

## 5. Mesh Alignment & Coordinate Frame Conversion

```python
outputs = torch.load("sam3d_object_cache_ptmp_input.pt")
```

```python
R_yup_to_zup = torch.tensor([[-1,0,0],[0,0,1],[0,1,0]], dtype=torch.float32)
R_flip_z = torch.tensor([[1,0,0],[0,1,0],[0,0,-1]], dtype=torch.float32)
R_pytorch3d_to_cam = torch.tensor([[-1,0,0],[0,-1,0],[0,0,1]], dtype=torch.float32)
R_rot_y_180 = torch.tensor([[-1, 0,  0], [ 0, 1,  0], [ 0, 0, -1]], dtype=torch.float32)
R_rot_z_180 = torch.tensor([[-1, 0,  0], [ 0, -1,  0], [ 0, 0, 1]], dtype=torch.float32)

def transform_mesh_vertices(vertices, rotation, translation, scale):

    if isinstance(vertices, np.ndarray):
        vertices = torch.tensor(vertices, dtype=torch.float32)

    vertices = vertices.unsqueeze(0)  #  batch dimension [1, N, 3]
    vertices = vertices @ R_flip_z.to(vertices.device) 
    # vertices = vertices @ R_rot_y_180.to(vertices.device)
    # vertices = vertices @ R_rot_z_180.to(vertices.device)
    vertices = vertices @ R_yup_to_zup.to(vertices.device)
    R_mat = quaternion_to_matrix(rotation.to(vertices.device))
    tfm = Transform3d(dtype=vertices.dtype, device=vertices.device)
    tfm = (
        tfm.scale(scale)
           .rotate(R_mat)
           .translate(translation[0], translation[1], translation[2])
    )
    vertices_world = tfm.transform_points(vertices)
    vertices_world = vertices_world @ R_pytorch3d_to_cam.to(vertices_world.device)
    
    return vertices_world[0]  # remove batch dimension


for i, out in enumerate(outputs):
    mesh = out["glb"]
    save_path = f"{PATH}/meshes/multi/{IMAGE_NAME}/demo_aligned/object_{i}.glb"
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    mesh.export(save_path)
    
    vertices = mesh.vertices
    vertices_tensor = torch.tensor(vertices)

    S = out["scale"][0].cpu().float()
    T = out["translation"][0].cpu().float()
    R = out["rotation"].squeeze().cpu().float()

    vertices_transformed = transform_mesh_vertices(vertices, R, T, S)
    mesh.vertices = vertices_transformed.cpu().numpy().astype(np.float32)
    mesh.fix_normals()

    save_path = f"{PATH}/meshes/multi/{IMAGE_NAME}/demo_aligned/object_{i}_aligned_rmat_tocam_flipz_ytoz.glb"
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    mesh.export(save_path)
```

## 4. Visualize Gaussian Splat of the Scene
### a. Animated Gif

```python
import trimesh
import time
import open3d as o3d
import pandas as pd
import torch.nn.functional as F


df_pixel_coord = pd.read_csv(f"{PATH}/../results/exp1/pixel_coords_debug.csv", header=None)
df_pixel_coord.columns = ['frame_id', 'object_id', 'tracklet_id', 'u', 'v']
df_3d_coord = pd.read_csv(f"{PATH}/../results/exp1/frontend_map_points_log.csv")

df = pd.merge(
    df_pixel_coord, 
    df_3d_coord, 
    on=['frame_id', 'object_id', 'tracklet_id'], 
    how='inner'
)
print(df.head())

frame_1 = df[df['frame_id'] == 1]

object_id = 1
geo = []
# Convert trimesh to Open3D
def trimesh_to_open3d(trimesh_mesh):
    import open3d as o3d
    import numpy as np

    mesh = o3d.geometry.TriangleMesh()
    mesh.vertices = o3d.utility.Vector3dVector(trimesh_mesh.vertices)
    mesh.triangles = o3d.utility.Vector3iVector(trimesh_mesh.faces)

    # Check for vertex colors in trimesh
    if hasattr(trimesh_mesh.visual, 'vertex_colors'):
        # Trimesh colors are often 0-255 RGBA; Open3D wants 0.0-1.0 RGB
        colors = trimesh_mesh.visual.vertex_colors[:, :3] / 255.0
        mesh.vertex_colors = o3d.utility.Vector3dVector(colors.astype(np.float64))
    
    return mesh

# 1. Transform your trimesh first
# mesh.apply_transform(model.params)
for i in [0, 1, 2, 3]:
    object_id = i+1
    mesh_original = trimesh.load(f'{PATH}/meshes/multi/{IMAGE_NAME}/demo_aligned/object_{i}.glb')
    mesh_aligned = trimesh.load(f'{PATH}/meshes/multi/{IMAGE_NAME}/demo_aligned/object_{i}_aligned_rmat_tocam_flipz_ytoz.glb')
    if isinstance(mesh_original, trimesh.Scene):
        mesh_original = mesh_original.dump(concatenate=True)
        mesh_aligned = mesh_aligned.dump(concatenate=True)
    
    
    # 2. Convert using the new function
    o3d_mesh_original = trimesh_to_open3d(mesh_original)
    o3d_mesh_aligned = trimesh_to_open3d(mesh_aligned)
    
    # 3. Proceed with visualization
    # o3d.visualization.draw_geometries([o3d_mesh, pcd_traj])
    o3d_mesh_original.compute_vertex_normals()
    o3d_mesh_aligned.compute_vertex_normals()
    # geo.append(o3d_mesh_original)
    geo.append(o3d_mesh_aligned)


# Create your trajectory points from DynoSAM
# pcd_traj = o3d.geometry.PointCloud()
# pcd_traj.points = o3d.utility.Vector3dVector(frame_1[['x_world', 'y_world', 'z_world']].values)
# pcd_traj.paint_uniform_color([1, 0, 0]) # Red trajectory
# geo.append(pcd_traj)

for i in range(4):
    points = outputs[i].get('pointmap')
    colors = outputs[i].get('pointmap_colors')
    binary_mask = masks[i]
    if isinstance(binary_mask, np.ndarray):
        binary_mask = torch.from_numpy(binary_mask)
    mask_rescaled = F.interpolate(
        binary_mask.unsqueeze(0).unsqueeze(0).float(), 
        size=(points.shape[0], points.shape[1]), 
        mode='nearest'
    ).squeeze()
    y_indices, x_indices = torch.where(mask_rescaled > 0)
    y_indices = y_indices.to(points.device)
    x_indices = x_indices.to(points.device)
    filtered_points = points[y_indices, x_indices]
    filtered_colors = colors[y_indices, x_indices]
    filtered_points[:, 0] = -filtered_points[:, 0]
    filtered_points[:, 1] = -filtered_points[:, 1]
    flat_points = filtered_points.reshape(-1, 3)
    flat_colors = filtered_colors.reshape(-1, 3)
    if flat_colors.max() > 1.0:
        flat_colors = flat_colors / 255.0
    valid_idx = ~(flat_points == 0).all(dim=1)
    final_points = flat_points[valid_idx]
    final_colors = flat_colors[valid_idx]
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(final_points)
    pcd.colors = o3d.utility.Vector3dVector(final_colors)
    geo.append(pcd)


origin = o3d.geometry.TriangleMesh.create_coordinate_frame(size=0.5)
geo.append(origin)



def safe_visualize(geometries):
    vis = o3d.visualization.Visualizer()
    vis.create_window(window_name="Alignment", width=800, height=600)
    
    for geom in geometries:
        vis.add_geometry(geom)
    
    # Manually update the renderer
    for _ in range(100):
        vis.update_geometry(None)
        vis.poll_events()
        vis.update_renderer()
        time.sleep(0.01)
    
    vis.run() # This starts the interaction
    vis.destroy_window()

safe_visualize(geo)
```
